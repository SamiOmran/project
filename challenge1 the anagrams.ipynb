```python
wordclean=sorted(list(set([word.strip().lower() for word in open('words.txt','r')])))
```

```python
scan=input()
```

```python
for curr_word in (wordclean):
    if len(scan)==len(curr_word) :
        scan_chars=sorted(list(scan))
        curr_word_chars=sorted(list(curr_word))
        
        if scan_chars==curr_word_chars:
             print(scan+" "+curr_word) 
  
          
    
```

```python
di_ct={}

for word in wordclean:
    simple=''.join(sorted(word)).lower()
    if simple in di_ct:
        di_ct[simple].append(word)
    else:
        di_ct[simple]=[word]
        
        
make_group=[]
for word in di_ct:
    if len(di_ct[word])>1:
        make_group.append(di_ct[word])
```

```python
di_ct
```

```python
di_ct_final={}
for anagrams in make_group:
    if len(anagrams[0])in di_ct_final:
        di_ct_final[len(anagrams[0])].append(anagrams)
        
    else:
        di_ct_final[len(anagrams[0])]=[anagrams]
```
